# Country Table

```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
import time
import pandas as pd
```

```python
urlpage = 'https://www.worldometers.info/coronavirus/#countries'
driver = webdriver.Chrome(r'C:/Users/susan/MyFolder/BDA/CovidBot/chromedriver.exe')
driver.get(urlpage)
html = driver.execute_script("return document.documentElement.outerHTML")
```

```python
table_list = pd.read_html(html)
```

```python
corona_table = table_list[0]
corona_table.iloc[:,1:]
```

```python
per = []
for i in range (0, len(corona_table)):
    a  = (corona_table.iloc[i]['TotalCases'].tolist()/corona_table.iloc[i]['Population'].tolist())*100
    per.append(a)
corona_table['Precentage'] = per
#corona_table
```

```python
corona_table.to_csv('C:/Users/susan/MyFolder/BDA/CovidBot/country.csv',index=False)
```
